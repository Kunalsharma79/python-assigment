10.  What is the difference between implicit and explicit type casting
 - The  difference is automation versus manual control:
Implicit Casting: Python automatically converts one type to another (e.g., adding an integer to a float) to prevent data loss without you writing extra code.
Explicit Casting: You manually force a conversion using functions like int(), float(), or str() to ensure the data is in the specific format you need.
